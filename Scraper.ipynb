```python
import requests

from lxml import html
from bs4 import BeautifulSoup

import pandas as pd
```

```python
headersparam = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.163 Safari/537.36"}

count = 0
_dict = {}
while True:
    if count == 10:
        break
    link = 'https://www.hepsiburada.com/laptop-notebook-dizustu-bilgisayarlar-c-98'

    r = requests.get(link, headers=headersparam)
    soup = BeautifulSoup(r.content, "html.parser", from_encoding='UTF-8')
    tree = html.fromstring(r.content)

    price = tree.xpath(f'//*[@id="i{count}"]/div/a/div[3]/div[1]/div[2]/text()')[0]
    model = tree.xpath(f'//*[@id="i{count}"]/div/a/div[3]/h3/text()')[0]
    link = 'https://www.hepsiburada.com' + soup.find('a', title=model)['href']

    r = requests.get(link, headers=headersparam)
    soup = BeautifulSoup(r.content, "html.parser", from_encoding='UTF-8')
    tree = html.fromstring(r.content)

    _count = 1
    # _list = []
    __dict = {}
    while True:
        feature = tree.xpath(f'//*[@id="productTechSpecContainer"]/table[2]/tbody/tr[{_count}]/th/text()')[0]
        values = tree.xpath(f'//*[@id="productTechSpecContainer"]/table[2]/tbody/tr[{_count}]/td/span/text()')
        
        if values != []:
            # _list.append({f'{feature}' : f'{values[0]}'})
            __dict[feature] = values[0]

        else:
            # _list.append({f'{feature}' : 'NaN'})
            __dict[feature] = 'NaN'
        _count += 1

        if _count == 10:
            count += 1
            # _dict[model] = _list
            _dict[model] = __dict
            break
```
